```python
#collecte de la donn√©es depuis yahoo finance

symbols = ['APPL', 'MSFT', 'AMZN', 'GOOG', 'META', 'TSLA', '^GSPC' ]

```

```python
#DATAFRAME

data_list = []

for i in symbols:
    data = webreader.DataReader('META', start = '2019-01-01', data_source='stooq')
    data['symbol'] = i
    data_list.append(data)
    
data_list
```

```python
#Visualisation des donn√©es avec pandas (Graphique)

for i in range(0, len(data_list)):
    data_list[i][['Close', 'Volume']].plot(subplots=True, title = data_list[i]['symbol'].unique()[0], figsize=(14,6)) 

plt.show()
```

```python
#DataFrame with all data (cr√©ation du dataframe)

df_all_data  = data_list[0];

for i in range(1, len(data_list)):
    df_all_data = pd.concat([df_all_data, data_list[i]])

df_all_data
```
